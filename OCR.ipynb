```python
from matplotlib import pyplot as plt
import numpy as np
import cv2
```

```python
image=cv2.imread('digits1.png')
```

```python
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
```

```python
plt.imshow(gray)
plt.title('image')
plt.show()
```

```python
divisions = list(np.hsplit(i,100) for i in np.vsplit(gray, 50))
```

```python
divisions_arr = np.array(divisions)
```

```python
divisions_arr.shape
```

```python
train = divisions_arr[:,:50].reshape(-1,400).astype(np.float32)
```

```python
train.shape
```

```python
test = divisions_arr[:,50:100].reshape(-1,400).astype(np.float32)
```

```python
test.shape
```

```python
from PIL import Image
```

```python
img = Image.open('digits1.png')
```

```python
img
```

```python
k = np.arange(10)
train_labels = np.repeat(k, 250)[:,np.newaxis]
test_labels = np.repeat(k, 250)[:,np.newaxis]
```

```python
knn = cv2.ml.KNearest_create()
```

```python
knn.train(train, cv2.ml.ROW_SAMPLE, train_labels)
```

```python
ret, output, neighbours, distance = knn.findNearest(test, k=5)
```

```python
int(output[0, 0])
```

```python
matched = output == test_labels
correct = np.count_nonzero(matched)
```

```python
accuracy = (correct*100)/(output.size)
```

```python
accuracy
```

```python
N7 = cv2.imread('1.png')
```

```python
plt.imshow(N7)
plt.title('image')
plt.show()
```

```python
N7_gray = cv2.cvtColor(N7, cv2.COLOR_BGR2GRAY)
```

```python
N7_gray_resized = cv2.resize(N7_gray, (20, 20))
```

```python
N7_flat = N7_gray_resized.reshape(-1, 400).astype(np.float32)
```

```python
ret, output, neighbors, distance = knn.findNearest(N7_flat, k=5)
```

```python
predicted_digit = int(output[0, 0])
```

```python
print("Predicted Digit:", predicted_digit)
```
